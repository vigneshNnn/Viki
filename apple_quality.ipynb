```python
import pandas as pd 
df = pd.read_csv("apple_quality.csv")
df
```

```python
df=df.dropna()
df
```

```python
df.isna().sum()
```

```python
df['Quality']=df['Quality'].replace(['good','bad'],[1,0])
```

```python
df
```

```python
x=df.drop(columns=['A_id','Quality'])
x
          
```

```python
y=df['Quality']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression,LogisticRegression,Lasso,Ridge
```

```python
m1=LinearRegression()
m1.fit(X_train,Y_train)
```

```python
m1.score(X_train,Y_train)
```

```python
m1.score(X_test,Y_test)
```

```python
m2=LogisticRegression()
m2.fit(X_train,Y_train)
m2
```

```python
m2.score(X_train,Y_train)
```

```python
m2.score(X_test,Y_test)
```

```python
m3=Lasso(alpha=50,max_iter=100,tol=0.1)
m3.fit(X_train,Y_train)
```

```python
m3.score(X_train,Y_train)
```

```python
m3.score(X_test,Y_test)
```

```python
m4=Ridge(alpha=50,max_iter=100,tol=0.1)
m4.fit(X_train,Y_train)
```

```python
m4.score(X_train,Y_train)
```

```python
m4.score(X_test,Y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
m5=DecisionTreeClassifier()
m5.fit(X_train,Y_train)
```

```python
m5.score(X_train,Y_train)
```

```python
m5.score(X_test,Y_test)
```
